```python
df['Delivery_person_Ratings'].fillna(df['Delivery_person_Ratings'].mean(), inplace=True)
```

```python
df['Delivery_person_Age'] = pd.to_numeric(df['Delivery_person_Age'], errors='coerce')
df['multiple_deliveries'] = pd.to_numeric(df['multiple_deliveries'], errors='coerce')
```

```python
import math

def distance(lat1, lon1, lat2, lon2):
    return math.sqrt((lat2-lat1)**2 + (lon2-lon1)**2)

df['Delivery_Distance'] = df.apply(
    lambda x: distance(
        x['Restaurant_latitude'],
        x['Restaurant_longitude'],
        x['Delivery_location_latitude'],
        x['Delivery_location_longitude']
    ), axis=1)
```

```python
df['Delivery_Speed'] = pd.cut(df['Time_taken(min)'],
                             bins=[0,20,40,60,120],
                             labels=['Fast','Normal','Slow','Very Slow'])
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))

    return R * c

df['Delivery_Distance_km'] = df.apply(
    lambda x: haversine(
        x['Restaurant_latitude'],
        x['Restaurant_longitude'],
        x['Delivery_location_latitude'],
        x['Delivery_location_longitude']
    ), axis=1
)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Delivery_Distance_km'], df['Time_taken(min)'])
plt.xlabel("Delivery Distance (km)")
plt.ylabel("Time Taken (minutes)")
plt.title("Distance vs Delivery Time")
plt.show()
```
